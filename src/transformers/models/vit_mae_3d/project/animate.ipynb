```python
from matplotlib import pyplot as plt
import numpy as np
from monai.data import ITKReader
from matplotlib import animation
import os


data_dir_prefix = '/media/minghui/Data/Datasets/NMSS Study/rchen/exp20230516/'
output_dir = './plots'

flair_lin_hm_files = {}
flair_nl_hm_files = {}
pd_lin_hm_files = {}
pd_nl_hm_files = {}
t1_lin_hm_files = {}
t1_nl_hm_files = {}
t2_lin_hm_files = {}
t2_nl_hm_files = {}

flair_lin_hm_files['baseline'] = '108-OSU-1_subject_108-1202_baseline_flair_mni_lin_hm.nii.gz'
flair_lin_hm_files['w24'] = '108-OSU-1_subject_108-1202_w24_flair_mni_lin_hm.nii.gz'
flair_lin_hm_files['w48'] = '108-OSU-1_subject_108-1202_w48_flair_mni_lin_hm.nii.gz'
flair_lin_hm_files['data_dir'] = 'agg_flair_lin_hm'

flair_nl_hm_files['baseline'] = '108-OSU-1_subject_108-1202_baseline_flair_mni_nl_hm.nii.gz'
flair_nl_hm_files['w24'] = '108-OSU-1_subject_108-1202_w24_flair_mni_nl_hm.nii.gz'
flair_nl_hm_files['w48'] = '108-OSU-1_subject_108-1202_w48_flair_mni_nl_hm.nii.gz'
flair_nl_hm_files['data_dir'] = 'agg_flair_nl_hm'

pd_lin_hm_files['baseline'] = '108-OSU-1_subject_108-1202_baseline_pd_mni_lin_hm.nii.gz'
pd_lin_hm_files['w24'] = '108-OSU-1_subject_108-1202_w24_pd_mni_lin_hm.nii.gz'
pd_lin_hm_files['w48'] = '108-OSU-1_subject_108-1202_w48_pd_mni_lin_hm.nii.gz'
pd_lin_hm_files['data_dir'] = 'agg_pd_lin_hm'

pd_nl_hm_files['baseline'] = '108-OSU-1_subject_108-1202_baseline_pd_mni_nl_hm.nii.gz'
pd_nl_hm_files['w24'] = '108-OSU-1_subject_108-1202_w24_pd_mni_nl_hm.nii.gz'
pd_nl_hm_files['w48'] = '108-OSU-1_subject_108-1202_w48_pd_mni_nl_hm.nii.gz'
pd_nl_hm_files['data_dir'] = 'agg_pd_nl_hm'

t1_lin_hm_files['baseline'] = '108-OSU-1_subject_108-1202_baseline_t1_mni_lin_hm.nii.gz'
t1_lin_hm_files['w24'] = '108-OSU-1_subject_108-1202_w24_t1_mni_lin_hm.nii.gz'
t1_lin_hm_files['w48'] = '108-OSU-1_subject_108-1202_w48_t1_mni_lin_hm.nii.gz'
t1_lin_hm_files['data_dir'] = 'agg_t1_lin_hm'

t1_nl_hm_files['baseline'] = '108-OSU-1_subject_108-1202_baseline_t1_mni_nl_hm.nii.gz'
t1_nl_hm_files['w24'] = '108-OSU-1_subject_108-1202_w24_t1_mni_nl_hm.nii.gz'
t1_nl_hm_files['w48'] = '108-OSU-1_subject_108-1202_w48_t1_mni_nl_hm.nii.gz'
t1_nl_hm_files['data_dir'] = 'agg_t1_nl_hm'

t2_lin_hm_files['baseline'] = '108-OSU-1_subject_108-1202_baseline_t2_mni_lin_hm.nii.gz'
t2_lin_hm_files['w24'] = '108-OSU-1_subject_108-1202_w24_t2_mni_lin_hm.nii.gz'
t2_lin_hm_files['w48'] = '108-OSU-1_subject_108-1202_w48_t2_mni_lin_hm.nii.gz'
t2_lin_hm_files['data_dir'] = 'agg_t2_lin_hm'

t2_nl_hm_files['baseline'] = '108-OSU-1_subject_108-1202_baseline_t2_mni_nl_hm.nii.gz'
t2_nl_hm_files['w24'] = '108-OSU-1_subject_108-1202_w24_t2_mni_nl_hm.nii.gz'
t2_nl_hm_files['w48'] = '108-OSU-1_subject_108-1202_w48_t2_mni_nl_hm.nii.gz'
t2_nl_hm_files['data_dir'] = 'agg_t2_nl_hm'


image_reader = ITKReader()
subject_id = '108-OSU-1_subject_108-1202'
# create output_dir if not exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
# data structure to save the images
images = {}

files_names = ['flair_lin', 'flair_nl', 'pd_lin', 'pd_nl', 't1_lin', 't1_nl', 't2_lin', 't2_nl']

for i, files in enumerate([flair_lin_hm_files, flair_nl_hm_files, pd_lin_hm_files, pd_nl_hm_files, t1_lin_hm_files, t1_nl_hm_files, t2_lin_hm_files, t2_nl_hm_files]):
    data_dir = files['data_dir']
    images[files_names[i]] = {}    
    # read the scans
    for scan_time in ['baseline', 'w24', 'w48']:
        # load the baseline image
        image = image_reader.read(os.path.join(data_dir_prefix, data_dir, files[scan_time]))
        # convert the list of images to a numpy array
        image = np.array(image)
        # crop to 182x182x182
        image = image[:, 18:200, :]
        # save in data structure
        images[files_names[i]][scan_time] = image
```

```python

# for scan_time in ['baseline', 'w24', 'w48']:

scan_time = 'baseline'
fig = plt.figure(figsize=(30, 80), dpi=25)
fig, axs = plt.subplots(3, 8)

def update(i):
    axs[0, 0].imshow(images['flair_lin'][scan_time][i, :, :])
    axs[0, 1].imshow(images['flair_nl'][scan_time][i, :, :])
    axs[0, 2].imshow(images['pd_lin'][scan_time][i, :, :])
    axs[0, 3].imshow(images['pd_nl'][scan_time][i, :, :])
    axs[0, 4].imshow(images['t1_lin'][scan_time][i, :, :])
    axs[0, 5].imshow(images['t1_nl'][scan_time][i, :, :])
    axs[0, 6].imshow(images['t2_lin'][scan_time][i, :, :])
    axs[0, 7].imshow(images['t2_nl'][scan_time][i, :, :])

    axs[1, 0].imshow(images['flair_lin'][scan_time][:, i, :])
    axs[1, 1].imshow(images['flair_nl'][scan_time][:, i, :])
    axs[1, 2].imshow(images['pd_lin'][scan_time][:, i, :])
    axs[1, 3].imshow(images['pd_nl'][scan_time][:, i, :])
    axs[1, 4].imshow(images['t1_lin'][scan_time][:, i, :])
    axs[1, 5].imshow(images['t1_nl'][scan_time][:, i, :])
    axs[1, 6].imshow(images['t2_lin'][scan_time][:, i, :])
    axs[1, 7].imshow(images['t2_nl'][scan_time][:, i, :])

    axs[2, 0].imshow(images['flair_lin'][scan_time][:, :, i])
    axs[2, 1].imshow(images['flair_nl'][scan_time][:, :, i])
    axs[2, 2].imshow(images['pd_lin'][scan_time][:, :, i])
    axs[2, 3].imshow(images['pd_nl'][scan_time][:, :, i])
    axs[2, 4].imshow(images['t1_lin'][scan_time][:, :, i])
    axs[2, 5].imshow(images['t1_nl'][scan_time][:, :, i])
    axs[2, 6].imshow(images['t2_lin'][scan_time][:, :, i])
    axs[2, 7].imshow(images['t2_nl'][scan_time][:, :, i])
    
    
    axs[0, 0].set_title('FLAIR Lin')
    axs[0, 1].set_title('FLAIR NL')
    axs[0, 2].set_title('PD Lin')
    axs[0, 3].set_title('PD NL')
    axs[0, 4].set_title('T1 Lin')
    axs[0, 5].set_title('T1 NL')
    axs[0, 6].set_title('T2 Lin')
    axs[0, 7].set_title('T2 NL')
    
    for i in range(3):
        for j in range(8):
            axs[i, j].axis('off')
    
    return axs

fig.tight_layout()
ani = animation.FuncAnimation(fig, update, frames=182, interval=100)
ani.save(os.path.join(output_dir, f'{subject_id}_{scan_time}.gif'), writer='imagemagick', fps=10, )
```
